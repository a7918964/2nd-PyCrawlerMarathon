### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

requests取回資料後，使用BeautifulSoup處理資料，資料型態是type(r.text)
```

```python
soup = BeautifulSoup(r.text, "html5lib")

print(soup)
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')

因為requests只是把資料攔截下來，還需要使用BeautifulSoup來解析資料，資料型態是type(soup)
```
